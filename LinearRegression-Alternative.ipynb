```python
import numpy as np
import pylab
```

```python
n = 100
theta_target = np.array([-45, 500]).reshape(2, 1)
alpha = np.array([0.0001, 0.0001]).reshape(2, 1)

E=np.random.normal(0, 100, n).reshape(n, 1)
X=np.ones(shape=(n, 2))
X[:, 0]=np.arange(n)
Y=X.dot(theta_target)+E
pylab.plot(X[:, 0], Y, 'o')
pylab.plot(X[:, 0], Y-E, 'k-', color='red')
pylab.show()
```

```python
theta = np.zeros(shape=(2, 1))

def get_gradient(theta, X, Y):
    gradient = np.zeros(len(theta)).reshape(len(theta),1)
    h = X.dot(theta)
    delta = h - Y
    sumdelta = delta.T.dot(X).T
    for i in xrange(gradient.size):
        gradient[i] = (1.0 / n) * sumdelta[i] * - 1
    gradient *= alpha
    return sumdelta, gradient

sumdelta = np.ones(len(theta)).reshape(len(theta),1)
while np.max(np.abs(sumdelta)) > 0.1:
    sumdelta, gradient = get_gradient(theta, X, Y)
    theta = theta + gradient
print theta
P=X.dot(theta)
pylab.plot(X[:, 0], Y, 'o')
pylab.plot(X[:, 0], Y-E, 'k-')
pylab.plot(X[:, 0], P, 'k-',  color='red')
pylab.show()
```
