```python
url="https://s3.amazonaws.com/hr-testcases/399/assets/trainingdata.txt"
import urllib2
import numpy as np
response = urllib2.urlopen(url)
training_data = np.array(map(lambda x:map(float, x.split(',')), response.read().split('\n')[:-1]))
training_data1 = np.array(filter(lambda x: x[0] < 4, training_data)).T #smaller than 4
training_data2 = np.array(filter(lambda x: x[0] >= 4, training_data)).T #larger than 4
```

```python
from scipy import stats
import pylab

training_data=training_data.T
x, y = training_data[0], training_data[1]

x1, y1 = training_data1[0], training_data1[1]
slope1, intercept1, r_value1, p_value1, slope_std_error1 = stats.linregress(x1, y1)

x2, y2 = training_data2[0], training_data2[1]
slope2, intercept2, r_value2, p_value2, slope_std_error2 = stats.linregress(x2, y2)

# Calculate some additional outputs
predict_y1 = intercept1 + slope1 * x1

pylab.plot(x, y, 'o')
pylab.show()

pylab.plot(x1, y1, 'o')
pylab.plot(x1, predict_y1, 'k-')
pylab.show()

predict_y2 = intercept2 + slope2 * x2


pylab.plot(x2, y2, 'o')
pylab.plot(x2, predict_y2, 'k-')
pylab.show()
```
