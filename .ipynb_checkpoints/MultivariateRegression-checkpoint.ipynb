```python
import numpy as np
import pylab
```

```python
n = 10
d = 4
base_X = 2
theta_target = np.array([-1, 5, 1, 50]).reshape(d, 1)
alpha = np.array([0.000001, 0.0001, 0.0001, 0.01]).reshape(d, 1)

E=np.random.normal(0, 50, n).reshape(n, 1)
X=np.ones(shape=(n, d))
X[:, 0]=np.arange(n)*np.arange(n)*np.arange(n)
X[:, 1]=np.arange(n)*np.arange(n)
X[:, 2]=np.arange(n)
Y=X.dot(theta_target)+E
pylab.plot(X[:, base_X], Y, 'o')
pylab.plot(X[:, base_X], Y-E, 'k-', color='red')
pylab.show()
```

```python
theta = np.zeros(shape=(d, 1))

def get_gradient(theta, X, Y):
    gradient = np.zeros(len(theta)).reshape(len(theta),1)
    h = X.dot(theta)
    delta = h - Y
    sumdelta = delta.T.dot(X).T
    for i in xrange(gradient.size):
        gradient[i] = (1.0 / n) * sumdelta[i] * - 1
    gradient *= alpha
    return sumdelta, gradient

sumdelta = np.ones(len(theta)).reshape(len(theta),1)
last_theta = np.ones(shape=(d, 1))
while np.max(np.abs(theta - last_theta)) > 0.001:
    sumdelta, gradient = get_gradient(theta, X, Y)
    last_theta = theta
    theta = theta + gradient

print theta
P=X.dot(theta)
pylab.plot(X[:, base_X], Y, 'o')
pylab.plot(X[:, base_X], Y-E, 'k-')
pylab.plot(X[:, base_X], P, 'k-',  color='red')
pylab.show()
```
