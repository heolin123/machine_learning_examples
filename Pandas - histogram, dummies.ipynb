```python
import pandas as pd
import statsmodels.api as sm
import pylab as pl
import numpy as np

df = pd.read_csv("http://www.ats.ucla.edu/stat/data/binary.csv")
df.columns = ["admit", "gre", "gpa", "prestige"]
print df.head()
pd.crosstab(df['admit'], df['prestige'], rownames=['admit'])
df.hist()
pl.show()
```
