```python
import numpy as np
import pylab
```

```python
n = 10
theta_target = np.array([-63, 665]).reshape(2, 1)
alpha = 0.01

E=np.random.normal(0, 86, n).reshape(n, 1)
X=np.ones(shape=(n, 2))
X[:, 0]=np.arange(n)
Y=X.dot(theta_target)+E
pylab.plot(X[:, 0], Y, 'o')
pylab.plot(X[:, 0], Y-E, 'k-', color='red')
pylab.show()
```

```python
theta = np.zeros(shape=(2, 1))
delta = np.ones(len(theta)).reshape(len(theta),1)
while np.max(np.abs(delta)) > 0.001:
    delta = np.zeros(len(theta)).reshape(len(theta),1)
    delta.astype('float64')
    for ind, x in enumerate(X):
        x = x.reshape(1, len(x))
        delta += (np.dot(x, theta) - Y[ind]) / n * alpha * x.T
    theta = theta - delta 

    
P=X.dot(theta)

pylab.plot(X[:, 0], Y, 'o')
pylab.plot(X[:, 0], Y-E, 'k-')
pylab.plot(X[:, 0], P, 'k-',  color='red')
pylab.show()
```
